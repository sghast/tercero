# METODOS ESCENCIALES NUMPY

### FUNDAMENTOS
| Método              | Qué hace              | Sintaxis                      |
| ------------------- | --------------------- | ----------------------------- |
| `np.array`          | crear array           | `np.array([1,2,3])`           |
| `np.zeros`          | array de ceros        | `np.zeros((f,c))`             |
| `np.ones`           | array de unos         | `np.ones((f,c))`              |
| `np.eye`            | matriz identidad      | `np.eye(n)`                   |
| `np.arange`         | rango con paso        | `np.arange(i,f,paso)`         |
| `np.linspace`       | valores equidistantes | `np.linspace(i,f,n)`          |
| `np.random.seed`    | fija aleatoriedad     | `np.random.seed(n)`           |
| `np.random.rand`    | floats [0,1)          | `np.random.rand(f,c)`         |
| `np.random.randint` | enteros aleatorios    | `np.random.randint(i,f,size)` |
| `reshape`           | cambiar forma         | `arr.reshape(f,c)`            |
| `flatten`           | aplanar               | `arr.flatten()`               |
| `np.sum`            | suma                  | `np.sum(arr)`                 |
| `np.mean`           | promedio              | `np.mean(arr)`                |
| `np.median`         | mediana               | `np.median(arr)`              |
| `np.std`            | desviación            | `np.std(arr)`                 |
| `np.min`            | mínimo                | `np.min(arr)`                 |
| `np.max`            | máximo                | `np.max(arr)`                 |
| `np.dot`            | producto punto        | `np.dot(a,b)`                 |
| `np.multiply`       | mult elemento         | `np.multiply(a,b)`            |
| `np.concatenate`    | unir arrays           | `np.concatenate([a,b])`       |

### ESTRUCTURA
| Método         | Qué hace            | Sintaxis            |
| -------------- | ------------------- | ------------------- |
| `np.ravel`     | aplana (vista)      | `arr.ravel()`       |
| `np.transpose` | transponer          | `arr.T`             |
| `astype`       | cambiar tipo        | `arr.astype(float)` |
| `copy`         | copia independiente | `arr.copy()`        |

| `np.maximum` | máximo elemento | `np.maximum(a,b)` |
| `np.minimum` | mínimo elemento | `np.minimum(a,b)` |


### FILTRADO
| Método             | Qué hace     | Sintaxis                |
| ------------------ | ------------ | ----------------------- |
