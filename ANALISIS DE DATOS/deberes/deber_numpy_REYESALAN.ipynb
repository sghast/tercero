### Estadísticas: `np.sum`, `np.mean`, `np.median`, `np.std`, `np.min`, `np.max`

Todas aceptan el parámetro `axis`:  
- `axis=0` → opera sobre filas (resultado por columna)  
- `axis=1` → opera sobre columnas (resultado por fila)  
- sin `axis` → opera sobre todo el array

````python
notas = np.array([[8.0, 7.5, 9.0],
                  [6.0, 8.5, 7.0],
                  [9.5, 6.5, 8.0]])

print('Media global:         ', np.mean(notas))
print('Media por evaluación: ', np.mean(notas, axis=0))  # por columna
print('Media por estudiante: ', np.mean(notas, axis=1))  # por fila
print('Std global:           ', round(np.std(notas), 4))
print('Min / Max:            ', np.min(notas), '/', np.max(notas))
print('Mediana:              ', np.median(notas))
print('Suma total:           ', np.sum(notas))
````

### `arr.ravel()` — aplanar a 1D (vista)

`ravel()` devuelve una **vista** del array aplanado: modificar el resultado puede modificar el original.  
`flatten()` (ya conocido) devuelve una **copia** independiente.

| Método | Tipo | Toca el original |
|---|---|---|
| `flatten()` | copia | no |
| `ravel()` | vista | sí |

````python
M = np.array([[1, 2, 3], [4, 5, 6]])
v = M.ravel()
print('ravel:', v)

v[0] = 99
print('Después de modificar v[0]=99, M[0,0]:', M[0, 0], '<-- cambió')
````

````python
# EJERCICIO 6
# Tienes registros de ventas en 3 semanas x 5 días.
# Necesitas la serie temporal completa (15 valores) para calcular la tendencia.

ventas_sem = np.array([
    [120, 95, 110, 105, 130],
    [115, 100, 125, 90, 140],
    [110, 105, 118, 122, 135]
])

# 6a. Aplana con ravel() para obtener la serie de 15 días
serie = ventas_sem.ravel()
print('Serie temporal:', serie)
print('Promedio del periodo:', round(np.mean(serie), 2))

# 6b. ¿ravel devuelve vista o copia? Verifica modificando serie[0]
serie[0] = 0
print('ventas_sem[0,0] después de modificar serie[0]:', ventas_sem[0, 0])
print('Conclusión: ravel devuelve', 'VISTA' if ventas_sem[0,0]==0 else 'COPIA')
````

### `np.transpose()` — intercambiar filas y columnas

Convierte un array `(m, n)` en `(n, m)`.  
Equivalente a `arr.T`.

**Caso de uso real:** los datasets llegan como `(features, muestras)` pero los modelos  
esperan `(muestras, features)`. Transpose lo corrige en una línea.

````python
A = np.array([[1, 2, 3], [4, 5, 6]])
print('A shape:', A.shape)
print(A)
print('\nA.T shape:', A.T.shape)
print(np.transpose(A))
print('\n¿A.T == np.transpose(A)?', np.all(A.T == np.transpose(A)))
````

````python
# EJERCICIO 7
# Dataset con 3 atributos y 5 muestras — shape (3, 5)
# Los algoritmos esperan (5, 3). Transpón y verifica.

datos = np.array([
    [1.2, 2.4, 0.8, 3.1, 1.9],   # feature 1
    [20,  35,  18,  42,  28 ],   # feature 2
    [0.5, 0.3, 0.8, 0.1, 0.6]   # feature 3
])

print('Shape original:', datos.shape)

# Transpón
datos_T = datos.transpose()
print('Shape transpuesto:', datos_T.shape)

# Primera muestra (fila 0 del transpuesto)
print('Primera muestra [f1, f2, f3]:', datos_T[0])
````

### `arr.copy()` — copia independiente

La asignación `b = a` crea un **alias**, no una copia.  
Modificar `b` modifica `a`. `copy()` crea un array completamente nuevo.

````python
original = np.array([10, 20, 30, 40, 50])

alias = original          # NO es copia
alias[0] = 999
print('original después de modificar alias:', original)

original2 = np.array([10, 20, 30, 40, 50])
copia = original2.copy()  # sí es copia
copia[0] = 999
print('original2 después de modificar copia:', original2)
````

### `np.greater`, `np.equal`, `np.maximum`, `np.minimum`

Son la versión funcional de los operadores `>`, `==`.  
Útiles cuando se pasan funciones como argumento o en pipelines.

- `np.greater(a, b)` → `a > b`  
- `np.equal(a, b)` → `a == b`

- `np.minimum(a, b)` → el menor de cada par

````python
a = np.array([1, 5, 3, 8, 2])
b = np.array([4, 2, 3, 1, 6])
print('a:', a)
print('b:', b)
print('greater(a,b):', np.greater(a, b))
print('equal(a,b):  ', np.equal(a, b))
print('maximum:     ', np.maximum(a, b))
print('minimum:     ', np.minimum(a, b))
````

### `np.where()` — if/else vectorizado

```python
np.where(condición, valor_si_verdadero, valor_si_falso)
```

Evalúa la condición elemento a elemento y selecciona uno de los dos valores.  
Es más eficiente que iterar con un `for`.

### `np.any`, `np.all`, `np.allclose`, `np.count_nonzero`

| Función | Pregunta |
|---|---|
| `np.any(cond)` | ¿hay al menos un True? |
| `np.all(cond)` | ¿son todos True? |
| `np.allclose(a, b)` | ¿son todos aproximadamente iguales? |

````python
stock = np.array([0, 5, 12, 0, 30, 0, 8])
print('¿Algún producto sin stock?  ', np.any(stock == 0))
print('¿Todos con stock?           ', np.all(stock > 0))
print('Productos disponibles:      ', np.count_nonzero(stock))

# allclose: para comparar resultados de cálculos numéricos (tolerancia de error)
a = np.array([1.0000001, 2.0000002])
b = np.array([1.0,       2.0      ])
print('¿Aprox iguales?             ', np.allclose(a, b))
````

## Datos faltantes y estadística robusta
Los datos reales siempre tienen valores faltantes representados como `NaN`.  
Las funciones estándar son **contagiadas** por NaN — usan las versiones `nan-safe`.

### `np.isnan`, `np.nanmean`, `np.nansum`, `np.nanstd`, `np.nanmin`, `np.nanmax`, `np.var`

El NaN es contagioso: `5 + NaN = NaN`.  
Las funciones `nan*` excluyen los NaN antes de calcular.

| Función | Propósito |
|---|---|
| `np.isnan(x)` | detecta posiciones NaN (devuelve máscara booleana) |
| `np.nanmean` / `np.nansum` / `np.nanstd` | estadísticas ignorando NaN |
| `np.nanmin` / `np.nanmax` | mín/máx ignorando NaN |
| `np.var(x)` | varianza (dispersión alrededor de la media) |

````python
datos = np.array([10.0, np.nan, 8.0, 9.5, np.nan, 7.0])
print('datos:', datos)
print('np.mean (contagiado): ', np.mean(datos))
print('np.nanmean (seguro):  ', np.nanmean(datos))

print('\nMáscara NaN:   ', np.isnan(datos))
print('Posiciones NaN:', np.where(np.isnan(datos))[0])

ventas = np.array([520.0, np.nan, 480.0, 610.0, np.nan, 495.0, 530.0])
print('\nEstadísticas nan-safe:')
print(f'  suma={np.nansum(ventas):.0f}  media={np.nanmean(ventas):.2f}  std={np.nanstd(ventas):.2f}')
print(f'  min={np.nanmin(ventas):.0f}   max={np.nanmax(ventas):.0f}')

# np.var
grupo_A = np.array([48, 50, 51, 49, 52])
grupo_B = np.array([20, 35, 50, 65, 80])
print(f'\nvar A={np.var(grupo_A):.2f}  var B={np.var(grupo_B):.2f}  (misma media, distinta dispersión)')
````

````python
# EJERCICIO 15
# Registro de temperaturas: 7 días x 3 sensores. Algunos fallaron (NaN).

temp = np.array([
    [22.1, np.nan, 21.8],
    [23.0, 22.5,   np.nan],
    [np.nan, 23.1, 22.9],
    [21.5, 22.0,   21.3],
    [24.1, np.nan, 23.8],
    [22.8, 22.1,   22.5],
    [21.0, 21.5,   np.nan]
])

# 15a. Total de lecturas faltantes
total_nan = np.sum(np.isnan(temp))
print('Lecturas faltantes:', total_nan)

# 15b. Temperatura promedio del periodo
prom = np.nanmean(temp)
print(f'Temperatura promedio: {prom:.2f} °C')

# 15c. Temperatura mínima y máxima del periodo
t_min = np.nanmin(temp)
t_max = np.nanmax(temp)
print(f'Rango: {t_min:.1f} — {t_max:.1f} °C')

# 15d. Varianza de las temperaturas válidas
temp_validas = temp[~np.isnan(temp)]
var = np.var(temp_validas)
print(f'Varianza: {var:.4f}')
````

````python
# EJERCICIO 16
# Comparar ventas reales con un presupuesto fijo de 500 por celda.

ventas_reales = np.array([
    [480, 510, 490],
    [520, 495, 505],
    [475, 530, 488]
])

# 16a. Crea la matriz de presupuesto con valor 500 (np.full)
presupuesto = np.full((3, 3), 500)

diferencia = ventas_reales - presupuesto

# 16b. Crea una máscara de ceros y marca con 1 donde no se cumplió la meta
bajo_meta = np.zeros_like(ventas_reales)
bajo_meta[diferencia < 0] = 1

print('Diferencia vs presupuesto:')
print(diferencia)
print('\nBajo meta (1 = no cumplió):')
print(bajo_meta)
````

### `np.append`, `np.insert`, `np.delete`, `np.unique`

Todas devuelven un **nuevo array** — NumPy no modifica arrays en lugar.

| Función | Descripción |
|---|---|
| `np.append(arr, values)` | agrega al final |
| `np.insert(arr, idx, values)` | inserta en una posición |
| `np.delete(arr, idx)` | elimina por índice |
| `np.unique(arr)` | valores únicos ordenados |

````python
arr = np.array([10, 20, 30, 40, 50])
print('original:           ', arr)
print('append 60:          ', np.append(arr, 60))
print('insert 25 en pos 2: ', np.insert(arr, 2, 25))
print('delete pos 1:       ', np.delete(arr, 1))

cats = np.array(['A', 'B', 'A', 'C', 'B', 'A', 'D'])
vals, counts = np.unique(cats, return_counts=True)
print('\nUnique + conteo:')
for v, c in zip(vals, counts):
    print(f'  {v}: {c} veces')
````

````python
# EJERCICIO 17
# Gestión del registro de IDs de productos.

ids = np.array([101, 102, 104, 105, 107])
print('IDs iniciales:', ids)

# 17a. El producto 103 fue omitido. Insértalo en el índice 2.
ids = np.insert(ids, 2, 103)
print('Después de insertar 103:', ids)

# 17b. El producto 107 fue descontinuado. Elimínalo (está al final).
ids = np.delete(ids, -1)
print('Después de eliminar 107:', ids)

# 17c. Llegó un lote con IDs [108, 109, 110]. Agrégalo al final.
ids = np.append(ids, [108, 109, 110])
print('Después de agregar lote:', ids)

# 17d. El sistema tiene duplicados. Obtén los IDs únicos.
ids_dup = np.append(ids, [103, 108, 101])
ids_unicos = np.unique(ids_dup)
print('IDs únicos:', ids_unicos)
````

### `np.random.uniform`, `np.random.randn`, `np.random.choice`

| Función | Distribución | Cuándo usarla |
|---|---|---|
| `uniform(low, high, size)` | Uniforme continua | valores entre dos límites con igual probabilidad |
| `randn(shape)` | Normal estándar (μ=0, σ=1) | datos con distribución gaussiana |
| `choice(arr, size, replace)` | Muestreo | seleccionar elementos de un conjunto |

Para obtener una normal con μ y σ propios: `μ + σ * np.random.randn(n)`

````python
np.random.seed(7)

# uniform
print('uniform [5, 50]:', np.round(np.random.uniform(5, 50, 5), 2))

# randn -> transformar a distribución real
alturas = 170 + 8 * np.random.randn(5)
print('alturas (μ=170, σ=8):', np.round(alturas, 1))

# choice
ciudades = np.array(['Quito', 'Guayaquil', 'Cuenca', 'Ambato'])
print('muestreo:', np.random.choice(ciudades, size=6, replace=True))
````

````python
# EJERCICIO 18
# Dataset sintético de 30 clientes.

np.random.seed(42)
n = 30

# 18a. Edades entre 18 y 65 (uniform, convertir a int)
edades = (np.random.uniform(18, 65, 30)).astype(int)

# 18b. Ingresos: media 800 USD, std 200 USD (randn)
ingresos = np.random.randn(30) * 200 + 800
ingresos = np.round(np.maximum(ingresos, 200), 2)  # mínimo 200

# 18c. Ciudad asignada por muestreo con reemplazo (choice)
ciudades = np.array(['Quito', 'Guayaquil', 'Cuenca', 'Ambato', 'Loja'])
ciudad_cliente = np.random.choice(ciudades, size=n, replace=True)

print('Primeros 5 clientes:')
print(f'{"Edad":>5}  {"Ingreso":>10}  Ciudad')
print('-' * 30)
for i in range(5):
    print(f'{edades[i]:>5}  {ingresos[i]:>10.2f}  {ciudad_cliente[i]}')

print(f'\nIngreso promedio: USD {np.mean(ingresos):.2f}')
print(f'Ciudades únicas: {np.unique(ciudad_cliente)}')
````

##  Persistencia: guardar y cargar arrays
Después de procesar datos, necesitamos guardarlos para reutilizarlos.  
- **`.npy`** (binario): rápido, exacto, solo NumPy.
- **`genfromtxt`** (texto): portable, legible por cualquier herramienta.

### `np.save` y `np.load` — formato binario `.npy`

Guarda el array con toda su precisión. Mucho más rápido que CSV  
para arrays grandes porque no convierte a texto.

````python
np.random.seed(1)
resultado = np.random.rand(5, 3)

np.save('resultado.npy', resultado)
print('Guardado. Tamaño en disco:', os.path.getsize('resultado.npy'), 'bytes')

cargado = np.load('resultado.npy')
print('¿Idéntico al original?', np.allclose(resultado, cargado))
````

### `np.genfromtxt` — leer archivos CSV con valores faltantes

Más robusto que `loadtxt`: maneja celdas vacías, tipos mixtos y cabeceras.

```python
np.genfromtxt('archivo.csv', delimiter=',', skip_header=1, filling_values=np.nan)
```

````python
# Creamos un CSV con valores faltantes para la demo
csv = """id,precio,stock
1,2.50,100
2,3.75,
3,1.20,200
4,,150
5,4.00,80
"""
with open('inventario.csv', 'w') as f:
    f.write(csv)

datos = np.genfromtxt('inventario.csv', delimiter=',',
                       skip_header=1, filling_values=np.nan)
print('Datos cargados:')
print(datos)
print('Faltantes por columna:', np.sum(np.isnan(datos), axis=0))
````

````python
# EJERCICIO 19 — Persistencia completa
# 1. Cargar CSV con faltantes
# 2. Imputar NaN con el promedio de la columna
# 3. Guardar limpio en .npy
# 4. Cargar y verificar

ventas_csv = """dia,norte,sur,centro
1,520,480,
2,,495,510
3,505,510,498
4,490,,520
5,530,488,505
"""
with open('ventas.csv', 'w') as f:
    f.write(ventas_csv)

# 19a. Carga ventas.csv
ventas = np.genfromtxt('ventas.csv', delimiter=',',skip_header=1, filling_values=np.nan)
print('Datos cargados:')
print(ventas)

# 19b. Imputa NaN con el promedio de cada columna de ventas (cols 1,2,3)
ventas_ok = ventas.copy()
for col in range(1, 4):
    columna = ventas[:, col]
    prom = np.nanmean(columna)
    ventas_ok[:, col] = np.where(np.isnan(columna), prom, columna)

print('\nDatos imputados:')
print(np.round(ventas_ok, 1))

# 19c. Guarda en ventas_ok.npy
np.save('ventas_ok.npy', ventas_ok)

# 19d. Carga y verifica que no hay NaN
check = np.load('ventas_ok.npy')
print('¿Quedan NaN?', np.any(np.isnan(check)))
````
