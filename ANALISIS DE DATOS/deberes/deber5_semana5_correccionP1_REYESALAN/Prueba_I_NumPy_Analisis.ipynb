El INAMI tiene un registro de temperaturas en grados Celsius del año 2026 para los meses de enero, febrero, marzo y abril de las capitales de provincia del Ecuador.

Use el archivo `temperaturas-ecuador.csv` como fuente de datos.

Primero cargue solo las columnas numéricas en una matriz llamada `temperaturas_ecuador` usando numpy.


````python
temperaturas_ecuador = np.genfromtxt('temperaturas-ecuador.csv',
                                      delimiter=',',
                                      skip_header=1,
                                      usecols=(1, 2, 3, 4)
                                      )
````

## B1. Diagnóstico inicial

2. Calcule suma y promedio global ignorando los NaN.


````python
print(f'Suma global: {np.nansum(temperaturas_ecuador).round(2)}')
print(f'Promedio global: {np.nanmean(temperaturas_ecuador).round(2)}')
````

````python
temperaturas_ecuador =np.round(temperaturas_ecuador, 2)
````

````python
print(f'Promedio por columna ignorando NaN:\n{np.nanmean(temperaturas_ecuador, axis=0).round(2)}')
````

## B3. Imputación simple

1. Crea una nueva matriz, llamada `matriz_limpia` y reemplace los NaN de cada columna por el **promedio de su propia columna**, utilice el promedio redondeado a dos decimales.


````python
matriz_limpia = np.where(np.isnan(temperaturas_ecuador), np.nanmean(temperaturas_ecuador, axis=0).round(2), temperaturas_ecuador)
````

Usa la siguiente url que contiene la canción WONDERWALL de OASIS:
https://raw.githubusercontent.com/juankgonzalez22/Prueba-I-Analisis/refs/heads/main/Wonderwall.txt


## C1. Limpieza básica

1. Mida el tiempo desde que empieza a realizar esta actividad.
2. Cargue el texto a una variable y pase todo a minúsculas, tome en cuenta que la canción solo es este fragmento de todo el txt(los metadatos de cabecera y pie de pagina deben omitirse en la carga de texto) .  
![imagen1](wonderwall.png)
3. Elimine signos de puntuación o caracteres especiales.
4. Tokenice el texto.


````python
url = 'https://raw.githubusercontent.com/juankgonzalez22/Prueba-I-Analisis/refs/heads/main/Wonderwall.txt'

# 1
inicioAB = time.time()
respuesta = request.urlopen(url)
texto = respuesta.read().decode('utf-8')
finA = time.time()
tiempo_ms = round((finA - inicioAB) * 1000, 3)

# 2
inicio_texto = texto.find('=== START OF THE SONG WONDERWALL BY OASIS ===') + 1
fin_texto = texto.find('=== END OF THE SONG WONDERWALL BY OASIS ===')
cancion = texto[inicio_texto:fin_texto]
cancion_minusculas = cancion.lower()

# 3
cancion_solo_letras = re.sub(r'[^a-z\s]', ' ', cancion_minusculas)

# 4
palabras = cancion_solo_letras.split()
````

## C2. Frecuencias

3. Filtre del texto solo palabras de longitud mayor a 4.
1. Obtenga las 5 palabras más frecuentes del texto previamente filtrado.
1. Muestre la cantidad de palabras únicas existente en toda la canción(sin haber aplicado el filtro de longitud mayor a 4).
1. Mida el tiempo de ejecución total hasta que terminó todo el proceso desde el inicio de C1, hasta que terminó la actividad 3 de C2.


````python
# 1
palabras_filtradas = [p for p in palabras if len(p) > 4]

# 2
contador = Counter(palabras_filtradas)
top_5 = contador.most_common(5)

print('PALABRA  FRECUENCIA')
for palabra, frecuencia in top_5:
    print(f'{palabra}:\t {frecuencia}')

# 3
unicas = len(Counter(palabras))
print(f'Cantidad de palabras únicas: {unicas}')

# 4
finB = time.time()
tiempo_s = round(finB - inicioAB, 3)
````

## C3. Reporte de texto

Genere un archivo de salida txt llamado `reporte_texto.txt`, este archivo debe decir:

- Total de palabras en todo el texto: ....
- Total de palabras únicas en todo el texto: ....
- Top 5 palabras más frecuentes de la canción(habiendo aplicado el filtro de palabras de longitud mayor a 4): .....
- Tiempo total en segundos (C1 A C2)


````python
with open('reporte_texto.txt', 'w', encoding='utf-8') as f:
    f.write(f'Total de palabras en todo el texto: {len(palabras)}\n')
    f.write(f'Total de palabras ÚNICAS en todo el texto: {unicas}\n')
    f.write('TOP 5 palabras más frecuentes\n')
    for palabra, frecuencia in top_5:
        f.write(f'{palabra}:\t {frecuencia}\n')
    
    f.write(f'Tiempo total en segundos (C1 A C2): {tiempo_s}s')
    
````

## D1.

Cree una matriz cuadrada `mi_matriz` de tamaño 80x80 con valores enteros aleatorios entre 10 y 500.


````python
dim = [80, 80]
mi_matriz = np.random.randint(10, 501, size=dim).astype(np.int64)
````

## D2.

Modifique la matriz para cumplir estas condiciones:

- La fila superior (primera fila) debe ser todo unos.
- La fila inferior (última fila) debe ser todo ceros.


````python
mi_matriz[0,:] = 1
mi_matriz[-1,:] = 0
print(mi_matriz)
````

## D3.

Calcule sobre la matriz final:

- Suma por filas
- Promedio por columnas
- Valor máximo y su posición


````python
sum_filas = np.sum(mi_matriz, axis=1)
print(f'Suma por filas:\n{sum_filas}')

mean_columnas = np.mean(mi_matriz, axis=0).round(2)
print(f'\nPromedio por columnas:\n{mean_columnas}')

print(f'\nValor máximo mi_matriz: {np.max(mi_matriz)}, posición: {np.argmax(mi_matriz)}')
print(f'Valor máximo sum_filas: {np.max(sum_filas)}, posición: {np.argmax(sum_filas)}')
print(f'Valor máximo mean_columnas: {np.max(mean_columnas)}, posición: {np.argmax(mean_columnas)}')
````

## D4.

Extraiga todos los elementos mayores o iguales a 300 y reporte:

- Cantidad
- Porcentaje respecto al total


````python
cantidad = np.count_nonzero(mi_matriz >= 300)
print(f'Cantidad elementos >= 300: {cantidad}')

porcentaje = ((cantidad/mi_matriz.size) * 100).round(2)
print(f'% respecto al total (6400): {porcentaje}%')
````

## D5.

Genere una matriz de clasificación ` usando la matriz final:

- "bajo" si valor <= 30
- "medio" si valor entre 200 y 300
- "alto" si valor >= 400


````python
clasificacion = mi_matriz.copy()

clasificacion = np.where(clasificacion <= 30, 'bajo', 
                         np.where((clasificacion >= 200) & (clasificacion <= 300), 'medio',
                                  np.where(clasificacion >= 400, 'alto', clasificacion)))

print(clasificacion)

bajos = np.sum(clasificacion == 'bajo')
medios = np.sum(clasificacion == 'medio')
altos = np.sum(clasificacion == 'alto')
bma = bajos + medios + altos

print('\nTOTALES')
print(f'> bajo: {bajos}')
print(f'> medio: {medios}')
print(f'> alto: {altos}')
print(f'> Sin clasificar: {mi_matriz.size - bma}')
````
