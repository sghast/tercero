### Ejercicio 3 — Slicing en 1D y 2D (2 pts)

Dado el siguiente array y matriz:

````python
# No modifiques esta celda, solo ejecútala
temperaturas = np.array([22.5, 24.1, 19.8, 26.3, 21.0, 28.7, 23.4, 20.6, 27.9, 25.2])
ventas = np.arange(1, 21).reshape(4, 5)  # ventas semanales por sucursal (4 sucursales, 5 semanas)
print('Temperaturas:', temperaturas)
print('Ventas:\n', ventas)
````

Ahora resuelve:

### Ejercicio 4 — Indexación booleana, fancy indexing y np.where (2 pts)

Sigue trabajando con el array `temperaturas` definido arriba.

a) Usa una máscara booleana para obtener solo las temperaturas mayores a 24.  
b) Usa fancy indexing para obtener las temperaturas en las posiciones [1, 4, 7].  
c) Usa `np.where` para crear un nuevo array donde las temperaturas mayores a 25 se mantengan iguales, y las menores o iguales se reemplacen por 0.  
d) Usa `np.where` para encontrar los **índices** de las temperaturas menores a 22.

````python
# --- Ejercicio 4: escribe tu código aquí ---

# a)
t_mayores_24 = temperaturas[temperaturas > 24]
print(f'temperaturas mayores a 24: {t_mayores_24}')

# b)
print(f'temperaturas en las posiciones [1, 4, 7]: {temperaturas[[1, 4, 7]]}')

# c)
nuevas_temp = np.where(temperaturas > 25, temperaturas, 0)
print(f'temperaturas menores a 25 reemplazadas por 0: {nuevas_temp}')

# d)
print(f'índices temperaturas menores a 22: {np.where(temperaturas < 22)[0]}')
````

````python
# No modifiques esta celda
np.random.seed(10)  # para que todos tengamos los mismos datos
calificaciones = np.random.randint(40, 100, size=(5, 4)).astype(float)
materias = ['Matemáticas', 'Física', 'Programación', 'Estadística']
estudiantes = ['Est_1', 'Est_2', 'Est_3', 'Est_4', 'Est_5']

print('Calificaciones (filas=estudiantes, columnas=materias):')
print(calificaciones)
````

a) Calcula el promedio general (de todos los valores).

## Parte 5 — Manejo de datos faltantes (NaN)

En datos reales es común tener valores que faltan. NumPy los representa con `np.nan`. El problema es que las funciones normales devuelven `nan` si hay aunque sea un valor faltante, por eso existen las funciones `nan*`.

### Ejercicio 8 — Trabajando con NaN (2 pts)

Vamos a simular que algunos estudiantes no rindieron ciertas materias. Metemos NaN en algunas posiciones de la matriz de calificaciones:

````python
# No modifiques esta celda
calif_con_nan = calificaciones.copy()  # copia para no dañar la original
calif_con_nan[0, 2] = np.nan   # Est_1 no rindió Programación
calif_con_nan[2, 0] = np.nan   # Est_3 no rindió Matemáticas
calif_con_nan[4, 3] = np.nan   # Est_5 no rindió Estadística

print('Calificaciones con datos faltantes:')
print(calif_con_nan)
````

a) Usa `np.isnan` para encontrar dónde están los valores faltantes. ¿Cuántos hay en total? (usa `np.sum` sobre el resultado de `isnan`).  
b) Intenta calcular `np.mean(calif_con_nan)`. ¿Qué resultado da? ¿Por qué?  
c) Ahora calcula la media correcta usando `np.nanmean`.  
d) Calcula `nanmin`, `nanmax` y `nanstd` de la matriz.  
e) Calcula el promedio por estudiante (por fila) ignorando los NaN.

````python
# --- Ejercicio 8: escribe tu código aquí ---
# a)
print(f'valores faltantes: {np.isnan(calif_con_nan).sum()}')

# b)
print(f'np.mean(calif_con_nan) da como resultado ´nan´ porque hay valores faltantes, solo hace falta uno en una cantidad N\n'
'de datos para que no calcule correctamente, para hacerlo bien se debe usar nanmean')

# c)
print(f'media correcta (np.nanmean): {np.nanmean(calif_con_nan).round(2)}')

# d)
print(f'minimo: {np.nanmin(calif_con_nan)}')
print(f'máximo: {np.nanmax(calif_con_nan)}')
print(f'desviación: {np.nanstd(calif_con_nan).round(2)}')

# e)
print(f'promedio por estudiante: {np.nanmean(calif_con_nan, axis=1).round(2)}')
````
