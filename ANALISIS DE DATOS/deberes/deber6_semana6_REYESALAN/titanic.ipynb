### 1.2 — Limpieza de datos

Identifica cuántos valores nulos hay por columna. ¿Qué columnas tienen más valores faltantes?

```python
print(titanic.isnull().sum().sort_values(ascending=False))
```

Imputa los valores nulos de la columna Age con la mediana de esa columna.

```python
titanic['Age'] = titanic['Age'].fillna(titanic['Age'].median())
```

Elimina la columna Cabin (tiene demasiados valores nulos para ser útil).

```python
titanic = titanic.drop(columns=['Cabin'])
titanic.head(5)
```

Elimina las filas que aún tengan valores nulos.

```python
titanic = titanic.dropna()
```

Verifica que no queden valores nulos usando .isnull().sum().

```python
display(titanic.isnull().sum())
```

### 1.3 — Transformación

Crea una nueva columna llamada FamilySize que sea la columna de SibSp + Parch + 1.

```python
titanic['FamilySize'] = titanic['SibSp'] + titanic['Parch'] + 1

titanic.head()
```

Crea una columna IsAlone que valga 1 si FamilySize == 1, y 0 en caso contrario.

```python
titanic['IsAlone'] = titanic['FamilySize'].apply(lambda x: 1 if x == 1 else 0)

titanic.head()
```

Convierte la columna Sex en valores numéricos: male = 0, female = 1.

```python
titanic['Sex'] = titanic['Sex'].apply(lambda x: 0 if x == 'male' else 1)

titanic.head()
```

### 1.4 — Análisis y preguntas

**P1. ¿Cuál fue la tasa de supervivencia general (porcentaje)?**

```python
tasa_supervivencia = titanic['Survived'].mean() * 100

print(f'Tasa de supervivencia general: {tasa_supervivencia:.2f}%')
```

*La tasa de supervivencia del 38.25%, para la época, nos indica que el crucero no contaba con los recursos suficientes para salvar más vidas y la arquitectura no era tan buena como parecía*

**P2. ¿Cuál es la tasa de supervivencia según el sexo del pasajero?**

```python
supervivencia_sexo = round(titanic.groupby('Sex')['Survived'].mean() * 100, 2)
print(f'{supervivencia_sexo}\nmale=0, female=1')
```

*Los resultados indican que la tasa de supervivencia de las mujeres fue mayor a la de los hombres. Se peude deber a la prioridad que se le dio a niños y mujeres, además del sacrificio de hombres por sus familias o parejas, como hizo Jack con Rose*

**P3. ¿Qué clase de pasajero (Pclass) tuvo la mayor tasa de supervivencia?**

```python
supervivencia_clase = round(titanic.groupby('Pclass')['Survived'].mean() * 100, 2)
print(f'{supervivencia_clase}\nalta=1, media=2, baja=3')
```

*Los pasajeros de primera clase son los que mayor tasa de supervivencia obtuvieron, una de las razpones puede ser que estaban mejor ubicados y tenían más información, además eran mejor tratados*

**P4. ¿Cuál es la edad promedio de los sobrevivientes vs. los no sobrevivientes?**

```python
supervivencia_edad = round(titanic.groupby('Survived')['Age'].mean(), 2)
supervivencia_edad.index = ['No sobrevivieron', 'Sobrevivieron']

print(supervivencia_edad)
```

*Aunque la diferencia no es tan grande, la prioridad fue para los más jóvenes y mujeres*

**P5. ¿Los pasajeros que viajaban solos (IsAlone == 1) tenían mayor o menor probabilidad de sobrevivir?**

```python
supervivencia_individual = round(titanic.groupby('IsAlone')['Survived'].mean() * 100, 2)
supervivencia_individual.index = ['Acompañados', 'Solos']

print(supervivencia_individual)
```

*En este caso si hay más diferencia, los resultados indican que era más probable sobrevivir si se viajaba acompañado, quizá porque no recibieron ayuda del que podría haber sido el/la acompañante*
