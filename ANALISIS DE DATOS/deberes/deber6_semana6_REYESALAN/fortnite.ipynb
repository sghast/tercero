### 2.1 — Exploración inicial

Muestra .head(), .tail(), .shape y .describe()

```python
fortnite = pd.read_csv('datasets/Fortnite Statistics.csv')

display(fortnite.head())
display(fortnite.tail())
display(fortnite.shape)
display(fortnite.describe())
```

Identifica cuántas columnas son numéricas y cuántas son categóricas

```python
col_numericas = fortnite.select_dtypes(include='number').columns
col_categoricas = fortnite.select_dtypes(include='object').columns

print(f'Columnas numéricas: {len(col_numericas)}')
print(f'Columnas categóricas: {len(col_categoricas)}')
```

Muestra el nombre y conteo de valores únicos de las columnas categóricas más importantes

```python
for col in col_categoricas:
    print(f'\nColumna: {col}')
    print(f'Valores únicos: {fortnite[col].nunique()}')
    print(fortnite[col].value_counts())
```

### 2.2 — Filtrado y selección

Selecciona únicamente las columnas numéricas del dataset y guárdalas en un nuevo DataFrame llamado df_fortnite_num

```python
df_fortnite_num = fortnite.select_dtypes(include='number')

display(df_fortnite_num)
```

Filtra y muestra solo las filas donde el jugador haya obtenido más de 5 kills

```python
mas_5_kills = df_fortnite_num[df_fortnite_num['Eliminations'] > 5]

display(mas_5_kills)
```

Filtra las partidas donde el jugador quedó en el top 10 (posición ≤ 10)

```python
top_10 = df_fortnite_num[df_fortnite_num['Placed'] <= 10]

display(top_10)
```

Usa .loc para mostrar la fila con el mayor número de kills en toda la historia del dataset

```python
max_kills = fortnite['Eliminations'].idxmax()

fortnite.loc[max_kills]
```

### 2.3 — Agrupación y agregación

Agrupa los datos por el tipo de modo de juego (si existe la columna) o por alguna columna categórica relevante

```python
grupos = fortnite.groupby('Mental State')
```

Calcula el promedio de kills, daño infligido y tiempo de partida por cada grupo

```python
promedios = round(grupos[['Eliminations', 'Damage to Players', 'Placed']].mean(), 2)

display(promedios)
```

Muestra el grupo con el mejor rendimiento promedio

```python
mejor_grupo = promedios['Eliminations'].idxmax()

print('Grupo con mejor rendimiento promedio:', mejor_grupo)
```

### 2.4 — Análisis y preguntas

**P1. ¿Cuál es el número máximo, mínimo y promedio de kills por partida?**

```python
maximo = fortnite['Eliminations'].max()
minimo = fortnite['Eliminations'].min()
promedio = fortnite['Eliminations'].mean()

print(f'Máximo: {maximo}')
print(f'Mínimo: {minimo}')
print(f'Promedio: {promedio:.2f}')
```

*El resultado indica que no es tan sencillo llegar a tener un número alto de kills, por lo que se debe tener práctica para lograrlo*

**P2. ¿Existe correlación entre el número de kills y el tiempo que el jugador sobrevive? Usa .corr() para responder**

![correlacion](correlacion.png)

```python
correlacion = fortnite['Eliminations'].corr(fortnite['Placed'])

print(f'Correlación: {correlacion:.5f}')
```

*Este resultado tiene sentido porque mientras más kills menor es el 'Placed' (tiempo de juego), no tiene correlación negativa perfecta (-1) porque no siempre se debe tener alguna/s kill/s para ganar una partida*

**P3. ¿En qué porcentaje de partidas el jugador terminó en el top 1 (ganó la partida)?**

```python
porcentaje_ganador = (fortnite['Placed'] == 1).mean() * 100

print(f'Porcentaje: {porcentaje_ganador:.2f}%')
```

*Este resultado indica que de las 87 partidas solo se ganaron el 3.45% (3), demostrando lo complicado que es quedar en primer lugar*
