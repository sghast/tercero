## Creación de Arrays


### `np.zeros()`, `np.ones()`, `np.full()` — Arrays con valores predefinidos

- `np.zeros()`: crea un arreglo lleno de ceros.
- `np.ones()`: crea un arreglo lleno de unos.
- `np.full()`: crea un arreglo lleno con un valor específico.

```python
display("Matriz de zeros:", np.zeros(5))
display("Matriz de ones:", np.ones((2, 3)))
display("Matriz de full:", np.full((2, 3), 7))
```

### `np.arange()` y `np.linspace()` — Secuencias numéricas

`np.arange(inicio, fin, paso)`
- Genera valores con un **paso fijo**.
- El valor final (`fin`) normalmente es **exclusivo** (no se incluye).
- Es ideal cuando te interesa controlar el incremento entre elementos.

`np.linspace(inicio, fin, num)`
- Genera una secuencia con una **cantidad fija de elementos** (`num`).
- Incluye el valor (`fin`) por defecto 
- Es ideal cuando necesitas repartir puntos de forma uniforme en un intervalo.

```python
# arange: inicio, fin (exclusivo), paso
print("Arage: ", np.arange(0, 10, 2))

# linspace: inicio, fin (inclusivo), cantidad de elementos
print("Linspace: ", np.linspace(0, 10, 5))
```

### `astype()` — Cambiar tipo de dato

Numéricas básicas

int → float (ej. arr.astype(float))

float → int (ej. arr.astype(int), que trunca decimales)

int32 ↔ int64 (dependiendo de la arquitectura o necesidad de precisión)

float32 ↔ float64 (para cálculos más precisos o más ligeros)

Booleanos

int → bool (0 se convierte en False, cualquier otro valor en True)

bool → int (False → 0, True → 1)

Strings

int/float → str (ej. arr.astype(str))

str → int/float (si las cadenas representan números válidos)

Tipos especiales

complex ↔ float/int (aunque al convertir de complejo a real se pierde la parte imaginaria)

datetime64 

```python
enteros = np.array([1, 2, 3])
flotantes = enteros.astype(np.float64)
cadena = enteros.astype(str)

print(enteros, enteros.dtype)
print(flotantes, flotantes.dtype)
print(cadena, cadena.dtype)
```

### `np.dot()` y `np.multiply()` —
## En vectores
`np.multiply`
Realiza multiplicación elemento a elemento (Hadamard product).
Cada posición se multiplica con la correspondiente del otro vector.


```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print('np.multiply:', np.multiply(a, b)) # -> 1*4 , 2*5 , 3*6
```

`dot` Calcula el producto escalar (inner product).Multiplica elemento a elemento y luego suma los resultados.



```python
print('np.dot:',np.dot(a, b))  # -> 1*4 + 2*5 + 3*6
```

```python
A = np.array([[1, 2],
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])

print('np.multiply:\n', np.multiply(A, B))
```

## Copias vs Vistas
### Diferencia fundamental entre `copy()` y una vista

```python
original = np.array([1, 2, 3, 4, 5])

# Vista: comparte memoria
vista = original[1:4]
vista[0] = 99
print('original (modificado por vista):', original)

# Copia: independiente
original2 = np.array([1, 2, 3, 4, 5])
copia = original2[1:4].copy()
copia[0] = 99
print('original2 (no modificado):', original2)
print('copia:', copia)
```


## Seccion 2: Tratamiento de Datos Faltantes con NumPy

La **limpieza de datos** es fundamental en cualquier análisis. En NumPy, los datos faltantes se representan como `np.nan` (_Not a Number_).

### 2.1 Crear y detectar valores NaN

Así es como se reserva un lugar para un elemento faltante en un array:


```python
array = np.array([1, 2, np.nan, 4, 5])
array
```

Para detectar la presencia de `NaN` usamos `np.isnan()`:

```python
np.isnan(array)
```

### 2.2 Cálculos con NaN

Las operaciones matemáticas estándar sobre arrays con `NaN` devuelven `NaN`:

```python
np.mean(array) #devuelve nan porque en el arreglo hay un dato faltante (NaN)
```

NumPy ofrece variantes de sus funciones que **ignoran los NaN** automáticamente:

```python
np.nanmean(array)
```

```python
print('suma sin NaN:', np.nansum(array))
print('mínimo sin NaN:', np.nanmin(array))
print('máximo sin NaN:', np.nanmax(array))
print('mediana sin NaN:', np.nanmedian(array))

```

### 2.3 Reemplazar valores NaN

Una estrategia común es sustituir `NaN` por otro valor usando `np.where()`, que funciona como un bucle condicional: _"donde se cumpla esta condición, haz X; donde no, haz Y"_.

**Reemplazar NaN por cero:**


```python
display(array)
array_con_0 = np.where(np.isnan(array), 0, array)
array_con_0
```

**Reemplazar NaN por el promedio del array:**

```python
display(array)
promedio = np.nanmean(array)
array_con_promedio = np.where(np.isnan(array), promedio, array)
array_con_promedio
```

### 2.4 Eliminar valores NaN

Otra estrategia es filtrar el array para **eliminar** los `NaN`. Usando `~` (tilde, que actúa como negación lógica) podemos conservar solo los valores válidos:

```python
# eliminar solo los NaN
array[np.isnan(array)]
```

```python
# sólo  los valores válidos
array_filtrado = array[~np.isnan(array)]
array_filtrado
```

> **Conclusión:** NumPy ofrece herramientas suficientes para identificar, manejar y sustituir valores `NaN`, lo que permite mantener la integridad de los análisis.

### 3.1 Importar datos con `genfromtxt()`

`genfromtxt()` es el método de NumPy para leer archivos de texto.Tiene particularidades importantes porque los arrays solo admiten **un tipo de dato** a la vez.

Si intentas leerlo sin parámetros adicionales, probablemente obtengas errores. Los parámetros esenciales son:

| Parámetro        | Descripción                                            |
| ---------------- | ------------------------------------------------------ |
| `delimiter`      | Caracter separador de columnas (ej: `','`)             |
| `skip_header`    | Número de filas iniciales a omitir (encabezados)       |
| `filling_values` | Valor con el que rellenar los datos faltantes          |
| `dtype`          | Tipo de dato del array resultante (ej: `int`, `float`) |

#### Paso 1 — Intentar leer el archivo sin configurar parámetros

Usaremos el archivo `medallas.csv`, que contiene el medallero olímpico. Si intentamos leerlo directamente, obtendremos un error porque el archivo tiene **texto y números mezclados**, y `genfromtxt()` necesita que le indiquemos cómo manejarlos.

```python
# generar un error
ruta = 'medallas.csv'
array_error = np.genfromtxt(ruta)
array_error
```

#### Paso 2 — Leer solo las columnas numéricas

Como los arrays solo admiten un tipo de dato, debemos **omitir la columna de texto** (País) y configurar correctamente los parámetros:


```python
ruta = 'medallas.csv'

array_medallas = np.genfromtxt(
    ruta,
    delimiter=',', # separador de columnas
    filling_values=0, # las celdas vacías se rellenan con 0
    skip_header=1, # omitir primera fila de títulos o encabezados
    dtype=int, #convertir los datos a enteros
    usecols=(0,1,2,3) # uso solo columnas numericas
)

print(f'La forma del array es: {array_medallas.shape}')
print(f'Primeras 5 filas (oro, plata, bronce, total):\n {array_medallas[:5]}')
```

Otra opcion seria sustituir los valores faltantes, por la media

```python
ruta = 'medallas.csv'

# cargar solo columnsa numericas y dejar faltantes como NaN(float)
array_medallas = np.genfromtxt(
    ruta,
    delimiter=',', # separador de columnas
    filling_values=np.nan, # las celdas vacías se rellenan con 0
    skip_header=1, # omitir primera fila de títulos o encabezados
    dtype=float, #convertir los datos a flotantes
    usecols=(0,1,2,3), # uso solo columnas numericas
    missing_values=''
)

# calcular la media por columna ignorando los NaN
medias_columnas = np.nanmean(array_medallas, axis=0)

# reemplaza cada NaN por la media de su columna
filas_nan, cols_nan = np.where(np.isnan(array_medallas))
array_medallas[filas_nan, cols_nan] = medias_columnas[cols_nan]

print('Medias por columna\n[Oro, Plata, Bronc, Total]\n', medias_columnas.round(2))

print('Primeras 5 filas imputadas por la media:')
print(array_medallas[:5])
```

#### Paso 3 — Explorar los datos importados con ufuncs

Una vez importados los datos como array, podemos aprovechar las ufuncs para obtener estadísticas rápidas:

```python
# Columnas: 0=Oro, 1=Plata, 2=Bronce, 3=Total

print("Total de medallas de ORO en el medallero:  ", np.trunc(np.sum(array_medallas[:,0])))
print("Total de medallas de PLATA en el medallero:", np.trunc(np.sum(array_medallas[:,1])))
print("Total de medallas de BRONCE en el medallero:", np.trunc(np.sum(array_medallas[:,2])))
print()
print("Promedio de medallas de oro por país:", np.round(np.mean(array_medallas[:,0]), 2))
print("País con más medallas totales (índice):", np.argmax(array_medallas[:,3]))
print("Máximo de medallas totales:            ", np.max(array_medallas[:,3]))
```

### 3.2 Exportar datos con `save()`

Guarda los datos en binario, lo que ocupa menos espacio y se carga más rápido que un archivo de texto plano.Esto esIdeal para matrices grandes (millones de elementos).

Para guardar un array en un archivo binario de NumPy usamos `save()`. Necesita dos argumentos principales:

- La **ruta de destino** (normalmente con extensión `.npy`)
- El **array** a exportar

Luego se recupera con `load()`, conservando forma y tipo de dato exactamente.

```python
array_ejemplo = np.array([[1, 2, 3],
                          [4, 5, 6]])

ruta_npy = 'mi_array_demo.npy'

# guardar en formato binario
np.save(ruta_npy, array_ejemplo)

# cargar el archivo creado
array_cargado = np.load(ruta_npy)
print('Array cargado correctamente')
print(array_cargado)
```
