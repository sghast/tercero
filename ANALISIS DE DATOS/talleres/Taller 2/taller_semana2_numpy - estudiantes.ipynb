## PARTE 1 — Construcción de Matrices Estructuradas

Construir matrices con **bordes y patrones específicos** usando slicing de NumPy.  
El patrón es siempre el mismo:

1. Crear la **base** (zeros u ones)
2. Sobreescribir los **bordes** con slicing


### Actividad 1 — Matriz M1: Bordes Aleatorios, Interior = 0(1 pt)

**Objetivo:** Crear una matriz 10×10 donde los **cuatro bordes tienen valores aleatorios** (1–10) y el **interior es cero**.

![Imagen referencial](https://i.imgur.com/trLQFG2.png)





````python
dim = (10, 10)

# PASO 1: Crear base de ceros
M1 = np.zeros(dim, dtype=int)

# PASO 2: Generar valores aleatorios para los bordes
aleatorios = np.random.randint(1, 11, 10)

# PASO 3: Asignar los cuatro bordes con slicing
M1[0,:] = aleatorios
M1[-1,:] = aleatorios
M1[:,0] = aleatorios
M1[:,-1] = aleatorios

print(M1)
````

**Reflexión A1:** ¿Qué pasa si usas `M1[1:-1, 1:-1]`? ¿Qué parte de la matriz accede?


````python
print('Accede a la parte central, a los zeros')
````

---

### Actividad 2 — Matriz M2: Bordes = 0, Interior Aleatorio(1 pt)

**Objetivo:** Crear la matriz **inversa a M1**: ahora el interior tiene valores aleatorios y los bordes son cero.  
![Imagen referencial](https://i.imgur.com/RiwlS39.png[/img])


````python
# PASO 1: Crear base con valores aleatorios en toda la matriz
M2_a = np.random.randint(1, 11, dim)

# PASO 2: Sobreescribir los cuatro bordes con cero
zeros_a = np.zeros(10, dtype=int)
M2_a[0,:] = zeros_a
M2_a[-1,:] = zeros_a
M2_a[:,0] = zeros_a
M2_a[:,-1] = zeros_a
print(f"FORMA A:\n{M2_a}\n")

'''
otra forma de hacerlo
paso 1: usar una matriz 10x10 de ceros
paso 2: crear una matriz 8x8 de aleatorios
paso 3: reemplazar con slicing la matriz de aleatorios en la matriz de ceros
'''

M2_b = np.zeros(dim, dtype=int)
aleatorios_b = np.random.randint(1, 11, [8, 8])
M2_b[1:-1, 1:-1] = aleatorios_b
print(f"FORMA B:\n{M2_b}")
````

Reto: No puntua, pero, como harias para hacer una matriz de ceros de (10x10) en donde solo en la diagonal van a ir numeros aleatorios?

````python
# con matriz identidad y multiplicar por los aleatorios
aleatorios_identidad = np.random.randint(1, 11, 10)

identidad = np.eye(10, dtype=int) * aleatorios_identidad
print(identidad)
````

## PARTE 2 — np.dot vs np.multiply




### ✏️ Actividad 4 — Calcular M3 = np.dot(M1, M2) y obten el maximo y minimo valor de la matriz M3(1 pt)


````python
# solo se usará M2_a
# uso de time.time() para el resultado de la actividad 6
inicio = time.time()
M3 = np.dot(M1, M2_a)
print('PARA M3:')
print(M3)
print(f'  valor mínimo: {M3.min()}')
print(f'  valor máximo: {M3.max()}')
fin = time.time()
duracion_dot = round((fin - inicio) * 1000, 5)
````

### ✏️ Actividad 5 — Calcular M4 = np.multiply(M1, M2) y obten el maximo y minimo valor de la matriz M3(1 pt)


````python
inicio = time.time()
M4 = np.multiply(M1, M2_a)
print('PARA M4:')
print(M4)
print(f'  valor mínimo: {M4.min()}')
print(f'  valor máximo: {M4.max()}')
fin = time.time()
duracion_mult = round((fin - inicio) * 1000, 5)
````

## PARTE 3 — Medición de Tiempos y Archivos .txt

Usa `time.time()` para medir el costo computacional de cada operación y guardaremos los resultados en archivos de texto.

**Patrón de medición:**

```python
inicio = time.time()
# ← operación a medir →
fin = time.time()
duracion = round((fin - inicio) * 1000, 5)  # en milisegundos
```


### ✏️ Actividad 6 — Medir Tiempos de np.dot y np.multiply usando M1 y M2 e indicar cual fue más rapido(1 pt)


````python
print(f'Duración en milisegundos de dot: {duracion_dot}')
print(f'Duración en milisegundos de multiply: {duracion_mult}')

if duracion_dot > duracion_mult:
    print('> Duracion_multiply fue más rápido')
else:
    print('> Duracion_dot fue más rápido')
````

Con matrices 10×10 los tiempos son muy pequeños (microsegundos). Para ver diferencias más claras, ejecutemos con matrices más grandes. Crea una Matriz de datos aleatorios de 1000 x 1000 (un millón de elementos)(no uses semilla) y vuelve a calcular el tiempo y realizar el dot y multiply y compara tiempos


````python
M_grande = np.random.randint(1, 500, [1000, 1000])

inicio_dot = time.time()
M_grande_dot = np.dot(M_grande, M_grande)
fin_dot = time.time()
duracion_dot_g = round(fin_dot - inicio_dot, 5)

inicio_mult = time.time()
M_grande_mult = np.multiply(M_grande, M_grande)
fin_mult = time.time()
duracion_mult_g = round(fin_mult - inicio_mult, 5)

print(f'Duración en segundos de dot: {duracion_dot_g}')
print(f'Duración en segundos de multiply: {duracion_mult_g}')

if duracion_dot_g > duracion_mult_g:
    print('> duracion_mult_g fue más rápido')
else:
    print('> duracion_dot_g fue más rápido')
````

### Ejercicio 9.2 — Imputar datos faltantes(1 pt)

9.2.1 .Crea una copia del array llamada `datos_limpios` donde:
- Los `NaN` en `Goles_campeon` (col 1) se reemplacen por la **mediana** de esa columna.
- Los `NaN` en `Goles_finalista` (col 2) se reemplacen por la **mediana** de esa columna.
- Los `NaN` en `Asistencia` (col 3) se reemplacen por el **promedio** de esa columna.

> 💡 Usa `np.nanmedian()`, `np.nanmean()` y `np.where()`.

````python
datos_limpios = np.copy(datos)
mediana_campeon = np.nanmedian(datos_limpios[:,1])
mediana_finalista = np.nanmedian(datos_limpios[:,2])
promedio_asistencia = np.nanmean(datos_limpios[:,3])

datos_limpios[:,1] = np.where(
    np.isnan(datos_limpios[:,1]),
    mediana_campeon,
    datos_limpios[:,1]
)

datos_limpios[:,2] = np.where(
    np.isnan(datos_limpios[:,2]),
    mediana_finalista,
    datos_limpios[:,2]
)

datos_limpios[:,3] = np.where(
    np.isnan(datos_limpios[:,3]),
    promedio_asistencia,
    datos_limpios[:,3]
)
````

9.2.2 Ejercicio  ( 1 pt)
Consulta sobre .npz en numpy y mandar a guardar datos_limpios en formato .csv incluyendo las cabeceras de cada columna y los datos

````python
'''
.npz es un archivo comprimido (como en zip de numpy) que guarda uno o varios arrays
puede usar np.savez() o np.savez_compressed()
'''

np.savez('datosfin.npz', datos=datos_limpios)

np.savetxt('datosfin.csv',
           datos_limpios,
           delimiter=',',
           header='Anio,Goles_campeon,Goles_finalista,Asistencia',
           fmt='%d'
           )
````
