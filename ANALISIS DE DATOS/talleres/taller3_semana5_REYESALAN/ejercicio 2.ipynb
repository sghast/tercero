## Introducción

En estos ejercicios vamos a trabajar con los datasets alojado en la ruta del curso **/Dia-05/datasets/XXVideos** representan el conjunto de estadísticas de vídeos de Youtube en los países de Canadá, USA y México.

Las columnas del dataframe significan lo siguiente:

1. `video_id` : Id del vídeo
2. `trending_date` : Fecha de registro de estadísticas
3. `title` : Título del vídeo
4. `channel_title` : Nombre del canal de Youtube
5. `category_id` : Id de la categoría del vídeo
6. `publish_time` : Fecha de publicación del vídeo
7. `tags` : Etiquetas del vídeo
8. `views` : Vistas del vídeo en la fecha 'trending_date'
9. `likes` : Likes del video en la fecha 'trending_date'
10. `dislikes` : Dislikes del video en la fecha 'trending_date'
11. `comment_count` : Número de comentarios en la fecha 'trending_date'
12. `thumbnail_link` : Link a la miniatura del vídeo
13. `comments_disabled` : Comentarios desactivados
14. `ratings_disabled` : Likes/Dislikes desactivados
15. `video_error_or_removed` : Video borrado o con error en subida
16. `description` : Descripción del vídeo

Ejecuta la siguiente celda para el dataframe con el contenido del dataset.

```python
import pandas as pd 

ca_videos = pd.read_csv("./datasets/CAvideos.csv")
mx_videos = pd.read_csv("./datasets/MXvideos.csv", encoding ='latin-1')
usa_videos = pd.read_csv("./datasets/usvideos.csv")
```

## Ejercicio 2 

Renombra las columnas `trending_date` y `publish_time` a los nombres `statistics_date` y `publish_date` respectivamente para los 3 paises.

```python
ca_videos = ca_videos.rename(columns={'trending_date': 'statistics_date', 'publish_time': 'publish_date'})
mx_videos = mx_videos.rename(columns={'trending_date': 'statistics_date', 'publish_time': 'publish_date'})
usa_videos = usa_videos.rename(columns={'trending_date': 'statistics_date', 'publish_time': 'publish_date'})

display(ca_videos.head())
display(mx_videos.head())
display(usa_videos.head())
```

## Ejercicio 3

Convierte las columnas `statistics_date` y `publish_date` de los 3 dataframe tipo de dato datetime64 con el formato **YYYY-MM-DD** para poder realizar funciones de agrupación y extracción de fechas.Consulta sobre strftime para esto


```python
ca_videos['statistics_date'] = pd.to_datetime(ca_videos['statistics_date'], format='%y.%d.%m')
mx_videos['statistics_date'] = pd.to_datetime(mx_videos['statistics_date'], format='%y.%d.%m')
usa_videos['statistics_date'] = pd.to_datetime(usa_videos['statistics_date'], format='%y.%d.%m')

ca_videos['publish_date'] = pd.to_datetime(ca_videos['publish_date'])
mx_videos['publish_date'] = pd.to_datetime(mx_videos['publish_date'])
usa_videos['publish_date'] = pd.to_datetime(usa_videos['publish_date'])

print(ca_videos.dtypes)
print(mx_videos.dtypes)
print(usa_videos.dtypes)
```

## Ejercicio 4

¿En que año se subieron más vídeos para cada país? 

**Ten en cuenta que puede haber más de una fila que compartan el mismo video (`video_id`)**

```python
# CANADÁ
videos_unicos_ca = ca_videos.drop_duplicates(subset='video_id')
anio_mas_videos_ca = videos_unicos_ca['publish_date'].dt.year.value_counts().idxmax()

# MEXICO
videos_unicos_mx = mx_videos.drop_duplicates(subset='video_id')
anio_mas_videos_mx = videos_unicos_mx['publish_date'].dt.year.value_counts().idxmax()

# ESTADOS UNIDOS
videos_unicos_usa = usa_videos.drop_duplicates(subset='video_id')
anio_mas_videos_usa = videos_unicos_usa['publish_date'].dt.year.value_counts().idxmax()
```

```python
print('Año con más videos publicados por país:')
print(f'\tCanadá: {anio_mas_videos_ca}')
print(f'\tMéxico: {anio_mas_videos_mx}')
print(f'\tEstados Unidos: {anio_mas_videos_usa}')
```

## Ejercicio 5 

¿Qué mes del año 2017 tiene el record en visitas para el dataframe de vídeos de México?

```python
videos_2017_mx = mx_videos[mx_videos['publish_date'].dt.year == 2017]

mes_mas_vistas = videos_2017_mx.groupby(videos_2017_mx['publish_date'].dt.month)['views'].sum().idxmax()
```

## Ejercicio 6

¿Qué dataframe de los 3 tiene más valores nulos entre todas sus columnas? Calcula el porcentaje de columnas nulas de los dataframes utilizando la función `apply`

```python
porcentaje_nulos_ca = ca_videos.apply(lambda col: col.isnull().mean() * 100).mean()
porcentaje_nulos_mx = mx_videos.apply(lambda col: col.isnull().mean() * 100).mean()
porcentaje_nulos_usa = usa_videos.apply(lambda col: col.isnull().mean() * 100).mean()

print(f'Porcentaje de nulos en CA: {round(porcentaje_nulos_ca, 2)}%')
print(f'Porcentaje de nulos en MX: {round(porcentaje_nulos_mx, 2)}%')
print(f'Porcentaje de nulos en USA: {round(porcentaje_nulos_usa, 2)}%')

if porcentaje_nulos_ca > porcentaje_nulos_mx and porcentaje_nulos_ca > porcentaje_nulos_usa:
    dataframe_mas_nulos = 'CA'
elif porcentaje_nulos_mx > porcentaje_nulos_usa:
    dataframe_mas_nulos = 'MX'
else:
    dataframe_mas_nulos = 'USA'

print(f'El dataframe con más nulos es: {dataframe_mas_nulos}')
```

## Ejercicio 7

Reemplaza los valores nulos del dataframe con más valores nulos con los siguientes criterios:

- 1. Si la columna es de tipo int64 reemplaza el valor nulo por la moda de valores de esa columna. (La moda es el valor que más se repite dentro de los valores de la columna.)
- 2. Si la columna es de tipo string rellena cada valor nulo por el valor no nulo más próximo después de ese registro. (Estrategia de relleno)

```python
df = mx_videos

for columna in df.columns:
    if df[columna].dtype == 'int64':
        moda_columna = df[columna].mode()[0]
        df[columna] = df[columna].fillna(moda_columna)
    else:
        df[columna] = df[columna].bfill()
```

## Ejercicio 8

¿Cuántos vídeos tienen audiciencia compartida entre los 3 dataframes? Es decir, que se hayan visto en USA, Canadá y México. 

**Puedes combinar los dataframes por la columna `video_id`**

```python
videos_comunes = set(ca_videos['video_id']) & set(mx_videos['video_id']) & set(usa_videos['video_id'])

print(f'Cantidad de videos con audiencia compartida: {len(videos_comunes)}')
```

## Ejercicio 9

¿Cuál es el vídeo más visto que se vió en los 3 países? ¿Y el que tuvo más likes?

```python
ca_comunes = ca_videos[ca_videos['video_id'].isin(videos_comunes)]
mx_comunes = mx_videos[mx_videos['video_id'].isin(videos_comunes)]
usa_comunes = usa_videos[usa_videos['video_id'].isin(videos_comunes)]

video_mas_visto_ca = ca_comunes.sort_values(by='views', ascending=False).iloc[0]
video_mas_visto_mx = mx_comunes.sort_values(by='views', ascending=False).iloc[0]
video_mas_visto_usa = usa_comunes.sort_values(by='views', ascending=False).iloc[0]

video_mas_likes_ca = ca_comunes.sort_values(by='likes', ascending=False).iloc[0]
video_mas_likes_mx = mx_comunes.sort_values(by='likes', ascending=False).iloc[0]
video_mas_likes_usa = usa_comunes.sort_values(by='likes', ascending=False).iloc[0]

print('VIDEO MÁS VISTO POR PAÍS')
print(f'\tCanadá: {video_mas_visto_ca["title"]}')
print(f'\tMéxico: {video_mas_visto_mx["title"]}')
print(f'\tUSA: {video_mas_visto_usa["title"]}')

print('\nVIDEO CON MÁS LIKES POR PAÍS')
print(f'\tCanadá: {video_mas_likes_ca["title"]}')
print(f'\tMéxico: {video_mas_likes_mx["title"]}')
print(f'\tUSA: {video_mas_likes_usa["title"]}')
```
